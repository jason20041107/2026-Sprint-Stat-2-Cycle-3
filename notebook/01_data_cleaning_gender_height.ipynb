# Data Cleaning Notebook: Gender and Height

## Project Cycle 3

This notebook is used for the data cleaning step of Project Cycle 3.

The selected research question is:

**Is the mean height different between male and female students?**

中文研究問題：

**男性與女性學生的平均身高是否存在差異？**

---

## Variables Used

In this project, we use two variables from the original `YRBS_2007.csv` dataset:

| Variable Type | Variable Name | Description |
|---|---|---|
| Group Variable | `WhatIsYourSex` | Student sex / gender group |
| Response Variable | `HowTallAreYouWithoutShoesInMeters` | Student height without shoes, measured in meters |

The group variable divides students into two groups:

1. Female students  
2. Male students  

The response variable is height, which is a quantitative variable. Therefore, this project will compare the mean height between male and female students.

---

## Data Cleaning Steps

The purpose of this notebook is to clean the original dataset and keep only the variables needed for the selected research question.

The main cleaning steps are:

1. Import required Python packages.
2. Read the original `YRBS_2007.csv` file from the raw data folder.
3. Select only the group variable and response variable.
4. Remove rows with missing values in sex or height.
5. Keep only valid sex codes.
6. Recode the sex variable into readable group names: `Female` and `Male`.
7. Create a simplified height variable named `Height_m`.
8. Save the cleaned dataset as a new CSV file in the processed data folder.

---

## Output File

The cleaned dataset will be saved as:

`YRBS_2007_cleaned_gender_height.csv`

This cleaned file will be used for later descriptive statistics, visualization, and Welch two-sample t-test analysis.

```python
from pathlib import Path
import pandas as pd

# 設定原始資料與輸出資料夾路徑
raw_dir = Path(r"C:\Users\curry\OneDrive\桌面\統計實習\project-cycle-3\data\data raw")
processed_dir = Path(r"C:\Users\curry\OneDrive\桌面\統計實習\project-cycle-3\data\data processed")

# 設定原始資料檔案名稱
raw_file = raw_dir / "YRBS_2007.csv"

# 如果輸出資料夾不存在，就自動建立
processed_dir.mkdir(parents=True, exist_ok=True)

# 讀取原始資料
df_raw = pd.read_csv(raw_file)

# 選定本研究需要的變數
group_var = "WhatIsYourSex"
response_var = "HowTallAreYouWithoutShoesInMeters"

# 只保留性別與身高兩個欄位
df_clean = df_raw[[group_var, response_var]].copy()

# 刪除性別或身高缺失值
df_clean = df_clean.dropna(subset=[group_var, response_var])

# 只保留有效性別編碼：1 = Female, 2 = Male
df_clean = df_clean[df_clean[group_var].isin([1, 2])]

# 保留合理身高資料，身高必須大於 0
df_clean = df_clean[df_clean[response_var] > 0]

# 將性別代碼轉換成文字，方便後續分析
df_clean["Sex"] = df_clean[group_var].map({
    1: "Female",
    2: "Male"
})

# 將身高欄位重新命名成較容易閱讀的欄位
df_clean["Height_m"] = df_clean[response_var]

# 重新整理欄位順序
df_clean = df_clean[[group_var, "Sex", response_var, "Height_m"]]

# 輸出清理後的資料
output_file = processed_dir / "YRBS_2007_cleaned_gender_height.csv"
df_clean.to_csv(output_file, index=False, encoding="utf-8-sig")

# 顯示清理結果
print("原始資料筆數：", df_raw.shape[0])
print("清理後資料筆數：", df_clean.shape[0])
print("各組樣本數：")
print(df_clean["Sex"].value_counts())

print("\n各組身高描述統計：")
print(df_clean.groupby("Sex")["Height_m"].agg(["count", "mean", "std", "min", "max"]))

print("\n清理後資料已輸出到：")
print(output_file)

# 顯示前五筆清理後資料
df_clean.head()
```
